```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
train_data = pd.read_csv('../data/logs_train.csv')
test_data = pd.read_csv('../data/logs_test.csv')

# Show the first few records
print(train_data.head())

# Load the dataset
df = pd.read_csv('../data/logs_train.csv')

# Display the first few rows of the DataFrame
df.head()

# Generate summary statistics
df.describe()

# Find missing values
df.isna().sum()

# Find missing values
df.isna().sum()

# Plotting histograms
df['Status'].hist()
plt.title('HTTP Status Codes Distribution')
plt.show()

# Boxplot
sns.boxplot(x=df['Status'])
plt.show()

# Kernel Density Plot
sns.kdeplot(df['Status'], fill=True)
plt.show()

# Convert timestamp to Unix time (assuming the timestamp column is named 'Timestamp')
df['Timestamp'] = pd.to_datetime(df['Timestamp']).astype(int) / 10**9

# Filter out only numerical columns
df_numeric = df.select_dtypes(include=['number'])

# Compute the correlation matrix
corr = df_numeric.corr()


# Generate a heatmap
sns.heatmap(corr, annot=True)
plt.show()

```
